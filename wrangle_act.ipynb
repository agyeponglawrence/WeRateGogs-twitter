```python
# Tweet IDs for which to gather additional data via Twitter's API
tweet_ids = archive.tweet_id.values
len(tweet_ids)
```

```python
# Query Twitter's API for JSON data for each tweet ID in the Twitter archive
count = 0
fails_dict = {}
start = timer()
# Save each tweet's returned JSON as a new line in a .txt file
with open('data/tweet_json.txt', 'w') as outfile:
    # This loop will likely take 20-30 minutes to run because of Twitter's rate limit
    for tweet_id in tweet_ids:
        count += 1
        print(str(count) + ": " + str(tweet_id))
        try:
            tweet = api.get_status(tweet_id, tweet_mode='extended')
            print("Success")
            json.dump(tweet._json, outfile)
            outfile.write('\n')
        except tweepy.errors.TweepyException as e:
            print("Fail")
            fails_dict[tweet_id] = e
            pass
end = timer()
print(end - start)
print(fails_dict)
```

```python
# Load the Json Data into a Data frame
df_list = []  # an empty list

# Read the Data and load into the empty dictionary
with open('data/tweet_json.txt') as json_file:
    for line in json_file:
        data = json.loads(line)
        tweet_id = data['id']
        retweet_count = data['retweet_count']
        favorite_count = data['favorite_count']

        df_list.append({
            'tweet_id': tweet_id,
            'retweet_count': int(retweet_count),
            'favorite_count': int(favorite_count)
        })

# Convert the list into a Dataframe
additional = pd.DataFrame(
    df_list, columns=['tweet_id', 'retweet_count', 'favorite_count'])
```

```python
# Create a function for checking missing values
def missing_values(dataframe):
    size = dataframe.shape[0]
    for column in dataframe.columns:
        missing_value_percent = 100 * sum(dataframe[column].isnull()) / size
        print(column +
              ': {}% missing values'.format(round(missing_value_percent, 2)))
        print('---' * 20)
```

### Research Question 3: *Based on rating, favorite counts and retweet counts, which breed is the best?*

To answer this question we will:
- Create an indicator (overall_rating) that combines the three metrics i.e., rating, favorite counts and retweet counts. 
- Then compare the indicator across the different breeds. 

```python
master[['rating', 'favorite_count', 'retweet_count']]
```

```python
# Create a function that normalizes the three variables to (0, 100)
def normalize(column):
    col_max = column.max()
    column = 100 * column / col_max
    return column
```

```python
# Normalize the columns
columns_for_indicator = ['rating', 'favorite_count', 'retweet_count']

for column in columns_for_indicator:
    master[column + '_norm'] = normalize(master[column])
```

```python
# Create the overall_rating indicator
overall_rating = (master.rating_norm + master.favorite_count_norm +
                  master.retweet_count_norm) / 3
master['overall_rating'] = overall_rating
```

```python
master.head()
```
